* ### Project Goal
> 오토인코더를 사용하여 흑백 이미지를 채색할 수 있는 모델을 만들어보자. 
* ### Methods
> 1. Train data의 RGB 이미지를 Lab color space로 변환
> 2. Lab로 변환한 train data에서 L 채널과 ab 채널 분리 
> 3. 분리한 채널 중 **L 채널을 input**으로 **ab채널을 Output**으로 하여 오토인코더로 훈련시킨다.
> 4. Test data의 L채널만 두고, ab채널은 훈련 웨이트 값으로 만들어 낸다. 
> 5. Lab를 다시 RGB로 변환한다.

### 1.1 Train Data Loading

```python

X_train_rgb = []

for idx, f in enumerate(train_files):
    try:
        # 이미지 파일 읽기
        img = cv2.imread(f)
        
        # 이미지를 224*224 크기로 리사이징
        img = cv2.resize(img, (224, 224))
        
        # cv2는 이미지를 BGR로 읽으므로 BGR->RGB전환
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # numpy array로 변환
        img = np.array(img)
        
        # X_train_rgb리스트에 추가
        X_train_rgb.append(img)
    
    except:
        print(str(i)+"file Error")
```

```python
X_train_rgb = np.array(X_train_rgb)
print(X_train_rgb.shape)
```

```python
# normalization
X_train_rgb = (X_train_rgb.astype(np.float32)) / 255.
```

```python
# RGB에서 LAB로 전환
# LAB 채널에서 L과 ab채널 분리
X_train_L = []
X_train_ab = []

for idx, img in enumerate(X_train_rgb):
    try:
        # RGB --> LAB
        Lab = cv2.cvtColor(img, cv2.COLOR_RGB2Lab)
        # L(Luminance)와 ab(color) 분리
        # L은 입력, ab는 label로 사용
        L = Lab[:, :, 0]
        ab = Lab[:, :, 1:] / 128.
        
        X_train_L.append(L)
        X_train_ab.append(ab)
    
    except:
        print("Error")
        
```

```python
X_train_L = np.array(X_train_L) 
X_train_ab = np.array(X_train_ab) 

# X_train_ab와 차원을 맞추기 위해 차원 추가 
X_train_L = np.expand_dims(X_train_L, -1) 
```

```python
print("shape of L: ", X_train_L.shape)
print("shape of ab: ", X_train_ab.shape)
```

## 2. Create the Model

```python
from keras.layers import *
from keras.models import Model, load_model
from keras.callbacks import ModelCheckpoint
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
from skimage.color import rgb2lab, lab2rgb
from skimage.transform import resize
from skimage.io import imsave
```

```python
en_input = Input(shape=(224, 224, 1,))
en = Conv2D(64, 3, activation='relu', padding='same', strides=2)(en_input)
en = Conv2D(128, 3, activation='relu', padding='same')(en)
en = Conv2D(128, 3, activation='relu', padding='same', strides=2)(en)
en = Conv2D(256, 3, activation='relu', padding='same')(en)
en = Conv2D(256, 3, activation='relu', padding='same', strides=2)(en)
en = Conv2D(512, 3, activation='relu', padding='same')(en)
en = Conv2D(512, 3, activation='relu', padding='same')(en)
en = Conv2D(256, 3, activation='relu', padding='same')(en)

de = Conv2D(128, 3, activation='relu', padding='same')(en)
de = UpSampling2D((2, 2))(de)
de = Conv2D(64, 3, activation='relu', padding='same')(de)
de = UpSampling2D((2, 2))(de)
de = Conv2D(32, 3, activation='relu', padding='same')(de)
de = Conv2D(16, 3, activation='relu', padding='same')(de)
de = Conv2D(2, 3, activation='tanh', padding='same')(de)
de_output = UpSampling2D((2, 2))(de)

model = Model(inputs=en_input, outputs=de_output)
```

```python
model.summary()
```

## 3. Train the Model

```python
# model compile
model.compile(optimizer='adam', loss='mse', metrics=['acc'])
```

```python

num_epoch = 100
batch_size = 32
```
